```python
import pandas as pd
pd_random = pd.read_csv('../data/output/random/IC_random_Epinions.csv')
pd_degreeDiscount = pd.read_csv('../data/output/degreeDiscount/IC_degreeDiscount_Epinions.csv')
pd_pageRank = pd.read_csv('../data/output/pageRank/IC_pageRank_Epinions.csv')
#pd_IMM = pd.read_csv('../data/output/IMM/IC_IMM_Epinions.csv')
pd_PMIA = pd.read_csv('../data/output/PMIA/IC_PMIA_Epinions.csv')
pd_CCA = pd.read_csv('../data/output/CCA/IC_CCA_Epinions.csv')
pd_KGC = pd.read_csv('../data/output/KGC/IC_KGC_f=0.2_Epinions.csv')
pd_RsetGreedy = pd.read_csv('../data/output/KGC/IC_RsetGreedy_Epinions.csv')
```

```python
# 运行时间对比
import pandas as pd
run_time = pd.DataFrame({'random':pd_random['run time'],
    'degreeDiscount':pd_degreeDiscount['run time'],
     'PageRank':pd_pageRank['run time'],
     'PMIA':pd_PMIA['run time'],
     'CCA':pd_CCA['run time'],
     'KGC':pd_KGC['run time'],
     'RsetGreedy':pd_RsetGreedy['run time'],})
run_time.index=[5,10,15,20,25,30,35,40,45,50]
ax = run_time.plot()
ax.set_xlabel('k')
ax.set_ylabel('run time')
ax.set_title('Comparison of running time')
fig = ax.get_figure()
fig.savefig('../data/output/figure/Epinions_run_time.png')


```
