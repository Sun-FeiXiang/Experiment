```python
import matplotlib.pylab as plt
import pandas as pd
from pandas import Series,DataFrame

# 直接应用算法的
pd_data_IC_degreeDiscount_hep_init = pd.read_csv('../data/output/IC_degreeDiscount_hep_init.csv')
pd_data_IC_degreeHeuristic_hep_init = pd.read_csv('../data/output/IC_degreeHeuristic_hep_init.csv')
pd_data_IC_generalGreedy_hep_init = pd.read_csv('../data/output/IC_generalGreedy_hep_init.csv')
pd_data_IC_newGreedyIC_hep_init = pd.read_csv('../data/output/IC_newGreedyIC_hep_init.csv')
pd_data_IC_randomHeuristic_hep_init = pd.read_csv('../data/output/IC_randomHeuristic_hep_init.csv')
pd_data_IC_singleDiscount_hep_init = pd.read_csv('../data/output/IC_singleDiscount_hep_init.csv')

# 应用k-core的
pd_data_IC_degreeDiscount_hep_k_core = pd.read_csv('../data/output/IC_degreeDiscount_hep.csv')
pd_data_IC_degreeHeuristic_hep_k_core = pd.read_csv('../data/output/IC_degreeHeuristic_hep.csv')
pd_data_IC_generalGreedy_hep_k_core = pd.read_csv('../data/output/IC_generalGreedy_hep.csv')
pd_data_IC_newGreedyIC_hep_k_core = pd.read_csv('../data/output/IC_newGreedyIC_hep.csv')
pd_data_IC_randomHeuristic_hep_k_core = pd.read_csv('../data/output/IC_randomHeuristic_hep.csv')
pd_data_IC_singleDiscount_hep_k_core = pd.read_csv('../data/output/IC_singleDiscount_hep.csv')

# 应用最小支配集的
pd_data_IC_degreeDiscount_hep_dominating_set = pd.read_csv('../data/output/IC_degreeDiscount_hep_dominating_set.csv')
pd_data_IC_degreeHeuristic_hep_dominating_set = pd.read_csv('../data/output/IC_degreeHeuristic_hep_dominating_set.csv')
pd_data_IC_generalGreedy_hep_dominating_set = pd.read_csv('../data/output/IC_generalGreedy_hep_dominating_set.csv')
pd_data_IC_newGreedyIC_hep_dominating_set = pd.read_csv('../data/output/IC_newGreedyIC_hep_dominating_set.csv')
pd_data_IC_randomHeuristic_hep_dominating_set = pd.read_csv('../data/output/IC_randomHeuristic_hep_dominating_set.csv')
pd_data_IC_singleDiscount_hep_dominating_set = pd.read_csv('../data/output/IC_singleDiscount_hep_dominating_set.csv')

pd_data_IC_CCA_hep = pd.read_csv('../data/output/IC_CCA_hep.csv')
pd_data_IC_CDCA_same_CCA_hep = pd.read_csv('../data/output/IC_CDCA_same_CCA_hep.csv')
pd_data_IC_CDCA_hep = pd.read_csv('../data/output/IC_CDCA_hep.csv')
pd_data_method2_hep = pd.read_csv('../data/output/IC_method2_hep.csv')
```

```python
# 运行时间对比
import pandas as pd
degreeDiscount_time = pd.DataFrame({'init':pd_data_IC_newGreedyIC_hep_init['run time'],
    'k_core':pd_data_IC_newGreedyIC_hep_k_core['run time'],
     'dominating set':pd_data_IC_newGreedyIC_hep_dominating_set['run time']})
degreeDiscount_time.index=[5,10,15,20,25,30]
ax = degreeDiscount_time.plot()
ax.set_xlabel('k')
ax.set_ylabel('run time')
ax.set_title('newGreedyIC')
fig = ax.get_figure()
fig.savefig('../data/output/figure/newGreedyIC_hep.png')
```

```python
# 覆盖个数对比
import pandas as pd
degreeDiscount_cover_size = pd.DataFrame({'init':pd_data_IC_newGreedyIC_hep_init['average cover size'],
    'k_core':pd_data_IC_newGreedyIC_hep_k_core['average cover size'],
     'dominating set':pd_data_IC_newGreedyIC_hep_dominating_set['average cover size']})
degreeDiscount_cover_size.index=[5,10,15,20,25,30]
ax = degreeDiscount_cover_size.plot()
ax.set_xlabel('k')
ax.set_ylabel('average cover size')
ax.set_title('newGreedyIC')
fig = ax.get_figure()
fig.savefig('../data/output/figure/newGreedyIC_hep_cover_size.png')
```

```python
# 覆盖大小对比
import pandas as pd
degreeDiscount_time = pd.DataFrame({'degreeDiscount':pd_data_IC_degreeDiscount_hep_init['average cover size'],
     'degreeHeuristic':pd_data_IC_degreeHeuristic_hep_init['average cover size'],
    'generalGreedy':pd_data_IC_generalGreedy_hep_init['average cover size'],
    'newGreedyIC':pd_data_IC_newGreedyIC_hep_init['average cover size'],
     'randomHeuristic':pd_data_IC_randomHeuristic_hep_init['average cover size'],
     'singleDiscount':pd_data_IC_singleDiscount_hep_init['average cover size'],
      'CCA':pd_data_IC_CCA_hep['average cover size'],
        'method1SameCCA':pd_data_IC_CDCA_same_CCA_hep['average cover size'],
        'method1':pd_data_IC_CDCA_hep['average cover size'],
        'method2':pd_data_method2_hep['average cover size']})
degreeDiscount_time.index=[5,10,15,20,25,30]
ax = degreeDiscount_time.plot()
ax.set_xlabel('k')
ax.set_ylabel('average cover size')
ax.set_title('all init algorithms average cover size')
fig = ax.get_figure()
fig.savefig('../data/output/figure/all_init_algorithms_hep_cover_size.png')


```
